# Trader algorithm

```python
from datamodel import OrderDepth, UserId, TradingState, Order
from typing import List
import string


class Trader:

    def __init__(self):
        splitting = [0.0, 0.5714285714285714, 0.2857142857142857, 0.14285714285714285, 0.0]
        self.q = {i + 1: splitting[i] for i in range(5)}
        self.quoting_percentage = 1.0
        self.position_limit = {"KELP": 50, "RAINFOREST_RESIN": 50}
        pass

    # Given a tradingstate, and a list of current_orders we already plan to execute, return whether or not we can add an additional order 'new_order'.
    def _can_add_order(self, state: TradingState, current_orders: list[Order], new_order: Order):
        max_possible_pos = state.position.copy()
        min_possible_pos = state.position.copy()

        if "RAINFOREST_RESIN" not in max_possible_pos:
            max_possible_pos["RAINFOREST_RESIN"] = 0
        if "RAINFOREST_RESIN" not in min_possible_pos:
            min_possible_pos["RAINFOREST_RESIN"] = 0

        for order in current_orders:
            if order.quantity < 0:
                min_possible_pos[order.symbol] += order.quantity
            elif order.quantity > 0:
                max_possible_pos[order.symbol] += order.quantity

        if new_order.quantity > 0:
            max_possible_pos[new_order.symbol] += new_order.quantity
        elif new_order.quantity < 0:
            min_possible_pos[new_order.symbol] += new_order.quantity

        for symbol, qty in max_possible_pos.items():
            if abs(qty) > self.position_limit[symbol]:
                return False

        for symbol, qty in min_possible_pos.items():
            if abs(qty) > self.position_limit[symbol]:
                return False

        return True

    def run(self, state: TradingState):
        result = {}
        for product in state.order_depths:
            orders: list[Order] = []

            if product == "RAINFOREST_RESIN":
                max_buy_amount = 50 - state.position[product]
                max_sell_amount = -50 - state.position[product]

            if product == "RAINFOREST_RESIN":
                # lets only buy at 9998 and only sell at 10002
                for i in [1, 2, 3, 4, 5]:
                    orders.append(Order(product, 10000 + i, int(self.quoting_percentage * max_sell_amount * self.q[i])))
                    orders.append(Order(product, 10000 - i, int(self.quoting_percentage * max_buy_amount * self.q[i])))

                # orders.append(Order(product, 10000 - self.k, 50 - state.position[product]))
                # orders.append(Order(product, 10000 + self.k, -50 - state.position[product]))
                # Some sizing scheme where I do 2 lots at each successive price level?
                # This means I show 2 at 10000 +- 1, 4 at 10000 +- 2, 6 at 10000 +- 3 etc etc.
                # Maybe 25 is too much and the entire order gets cancelled
                # for i in range(1, 5 + 1, 1):

                #     if self._can_add_order(state, orders, Order(product, 10000 + i, -self.k * i)):
                #         orders.append(Order(product, 10000 + i, int(-self.k * i)))
                #     if self._can_add_order(state, orders, Order(product, 10000 - i, self.k * i)):
                #         orders.append(Order(product, 10000 - i, int(self.k * i)))

                # desired_position = self.calculate_desired_position(state)
                # current_position = state.position['RAINFOREST_RESIN']

            result[product] = orders
        traderData = "SAMPLE"  # String value holding Trader state data required. It will be delivered as TradingState.traderData on next execution.

        conversions = 1
        return result, conversions, traderData
```

```python
from joblib import Parallel, delayed
from tqdm import tqdm

listings = {
    "KELP": Listing(symbol="KELP", product="KELP", denomination="SEASHELLS"),
    "RAINFOREST_RESIN": Listing(symbol="RAINFOREST_RESIN", product="RAINFOREST_RESIN", denomination="SEASHELLS"),
}



position_limit = {
    "KELP": 50,
    "RAINFOREST_RESIN": 50,
}



def calc_rainforest_resin_fair(order_depth: OrderDepth) -> float:


    return 10000



fair_value_evaluator = {
    # omit dictionary entry for kelp, so that it uses default behaviour (best_bid+best_ask)/2
    "RAINFOREST_RESIN": calc_rainforest_resin_fair,
}



market_data = pd.read_csv(os.path.join("..", "backtester", "market_data.csv"), sep=";")


trade_history = pd.read_csv(os.path.join("..", "backtester", "trade_history.csv"), sep=";")



def recurse(answer: list[list[int]], current_bucket: list[int], buckets: int, bucket_idx: int, balls_left: int):
    if bucket_idx == buckets:
        if balls_left == 0:
            answer.append(current_bucket.copy())
        return

    for i in range(balls_left + 1):
        current_bucket.append(i)
        recurse(answer, current_bucket, buckets, bucket_idx + 1, balls_left - i)
        current_bucket.pop()


# stars and bars / buckets and balls
def all_splittings(buckets: int, balls: int) -> list[list[int]]:
    answer = []
    recurse(answer, [], buckets, 0, balls)
    return answer


fineness = 5
splittings = all_splittings(5, fineness)
for splitting in splittings:
    for i in range(len(splitting)):
        splitting[i] /= fineness


quoting_percentages = np.linspace(0, 1, 10)


configs = []
for splitting in splittings:
    for qp in quoting_percentages:
        configs.append({"splitting": splitting, "quoting_percentage": qp})


def run_backtest(config):
    trader = Trader()
    trader.q = {i + 1: config["splitting"][i] for i in range(5)}
    trader.quoting_percentage = config["quoting_percentage"]
    bt = Backtester(
        trader=trader,
        listings=listings,
        position_limit=position_limit,
        fair_value_evaluator=fair_value_evaluator,
        market_data=market_data,
        trade_history=trade_history,
        spread_crossing_warning=False,
    )
    bt.run()
    pnl_resin = bt.pnl["RAINFOREST_RESIN"]
    return (config, pnl_resin)


results = Parallel(n_jobs=-1)(delayed(run_backtest)(config) for config in tqdm(configs, desc="Backtesting"))

results = sorted(results, key=lambda x: x[1], reverse=True)

for config, pnl in results:
    print(f"{config['splitting']}, {config['quoting_percentage']}, pnl = {pnl}")
```

It makes most sense to try a market making strategy because the fair price is fixed at 10000  

Lets look into something called the Stoikov model.  


The stoikov model helps balance two goals:  
1. Earn profits from bid ask spread
2. Avoid risk from holidng too much inventory


Something to consider is that the main risk from marketmaking is if the price of the underlying asset fluctuates. But, we know that the price of the asset here is constant, so does this simplify our model a lot?   

Before getting into the complicated MM model lets try create some simple model.
We want output of (price, quantity) to quote at.  


We know the fair price is exactly 10000. in MM, the profit is |x - 10000|, where x is the price you execute at (assuming you're executing at the correct side).  


Suppose f(x) is the frequency at which we execute a trade at price x over the entire time period, we therefore want to maximise f(x)|x-10000|.  

Can we derive something for f? we can try to model it with an exponential distribution because from the graph it looks like these rates are quite evenly spread out and using an exp distribution makes it quite easy.  

Probably not necessary actually. Can just gridsearch.  


Lets not worry about determining strategy based on current price because the price fluctuates so often anyways, just have a fixed strategy of how much to quote on each side.   
trades at $\Delta 2$ happen twice as often as trades at $\Delta 4$ which are about same frequence as at $\Delta 5$.? Initial eyeball guess.  


Lets have 3 parameters, $q_1$, $q_2$, $q_3$ for the percentages of maximum to quote at $\Delta 2,4,5$. 

```python
# Ensure the project root is in PATH.
import sys

sys.path.append("../")
# All imports of our code are relative to the project root.


from backtester.backtester import Backtester
from backtester.datamodel import TradingState, OrderDepth, Order, Listing

import numpy as np
import pandas as pd
import sys
import os

from datamodel import OrderDepth, UserId, TradingState, Order
from typing import List
import string


class Trader1:
    def __init__(self):
        self.position_limit = {"KELP": 50, "RAINFOREST_RESIN": 50}
        pass

    def take_best_orders(self, state: TradingState, orderbook: OrderDepth) -> list[Order]:
        orders: list[Order] = []

        symbol = "RAINFOREST_RESIN"
        position = state.position[symbol] if symbol in state.position else 0

        max_buy_amount = self.position_limit[symbol] - position
        max_sell_amount = abs(-self.position_limit[symbol] - position)

        if len(orderbook.buy_orders) != 0:
            best_bid_price = max(orderbook.buy_orders.keys())
            best_bid_volume = orderbook.buy_orders[best_bid_price]

            if best_bid_price > 10000:
                fill_quantity = min(max_sell_amount, best_bid_volume)

                if fill_quantity > 0:
                    orders.append(Order(symbol, best_bid_price, -fill_quantity))
                    del orderbook.buy_orders[best_bid_price]

        if len(orderbook.sell_orders) != 0:
            best_ask_price = min(orderbook.sell_orders.keys())
            best_ask_volume = abs(orderbook.sell_orders[best_ask_price])

            if best_ask_price < 10000:
                fill_quantity = min(max_buy_amount, best_ask_volume)

                if fill_quantity > 0:
                    orders.append(Order(symbol, best_ask_price, fill_quantity))
                    del orderbook.sell_orders[best_ask_price]

        return orders

    def run(self, state: TradingState):
        result = {}
        for product in state.order_depths:
            orders: list[Order] = []

            if product == "RAINFOREST_RESIN":
                took = self.take_best_orders(state, state.order_depths[product])

                while len(took) != 0:
                    orders = orders + took
                    took = self.take_best_orders(state, state.order_depths[product])

            result[product] = orders
        traderData = "SAMPLE"  # String value holding Trader state data required. It will be delivered as TradingState.traderData on next execution.

        conversions = 1
        return result, conversions, traderData
```

```python
from joblib import Parallel, delayed
from tqdm import tqdm

listings = {
    "KELP": Listing(symbol="KELP", product="KELP", denomination="SEASHELLS"),
    "RAINFOREST_RESIN": Listing(symbol="RAINFOREST_RESIN", product="RAINFOREST_RESIN", denomination="SEASHELLS"),
}


position_limit = {
    "KELP": 50,
    "RAINFOREST_RESIN": 50,
}


def calc_rainforest_resin_fair(order_depth: OrderDepth) -> float:

    return 10000


fair_value_evaluator = {
    # omit dictionary entry for kelp, so that it uses default behaviour (best_bid+best_ask)/2
    "RAINFOREST_RESIN": calc_rainforest_resin_fair,
}


market_data = pd.read_csv(os.path.join("..", "backtester", "market_data.csv"), sep=";")


trade_history = pd.read_csv(os.path.join("..", "backtester", "trade_history.csv"), sep=";")


trader = Trader1()
bt = Backtester(
    trader=trader,
    listings=listings,
    position_limit=position_limit,
    fair_value_evaluator=fair_value_evaluator,
    market_data=market_data,
    trade_history=trade_history,
    output_log_filename="trader1.log",
    spread_crossing_warning=False,
)

bt.run()
print(f"PNL: {bt.pnl}")

bt._log_trades(bt.output_log_filename)
```
